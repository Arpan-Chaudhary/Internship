```python
# Top 10 ODI teams in men’s cricket along with the records for matches, points and rating
```

```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
```

```python
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
team = soup.find('span',class_="u-hide-phablet")
team.text
```

```python
teams = []

for i in soup.find_all('span',class_="u-hide-phablet"):
    teams.append(i.text)
    
teams
```

```python
teams = [teams[0],teams[1],teams[2],teams[3],teams[4],teams[5],teams[6],teams[7],teams[8],teams[9]]

teams
```

```python
match = soup.find('td',class_="rankings-block__banner--matches")

match.text
```

```python
point = soup.find('td',class_="rankings-block__banner--points")

point.text
```

```python
matches = []

for i in soup.find_all('td',class_="table-body__cell u-center-text"):
    matches.append(i.text)
    
matches
```

```python
len(matches)
```

```python
a = matches
```

```python
a
```

```python
l1 = [match.text,a[0],a[2],a[4],a[6],a[8],a[10],a[12],a[14],a[16]]
```

```python
l1
```

```python
l2 = [point.text,a[1],a[3],a[5],a[7],a[9],a[11],a[13],a[15],a[17]]
l2
```

```python
rating = []

for i in soup.find_all("td",class_="rankings-block__banner--rating u-text-right"):# first place team ratings
    rating.append(i.text.replace("\n",""))
    
rating
```

```python
ratings = []

for i in soup.find_all('td',class_="table-body__cell u-text-right rating"):
    ratings.append(i.text)
    
ratings
```

```python
l5 = rating

l5
```

```python
l6 = [ratings[0],ratings[1],ratings[2],ratings[3],ratings[4],ratings[5],ratings[6],ratings[7],ratings[8]]

l6
```

```python
l5.extend(l6)
```

```python
l5
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({'Team':teams,'Matches':l1,'Points':l2})

df
```

```python
df1 = pd.DataFrame({'Rating':l5})

df1
```

```python
df =  pd.concat([df,df1],axis = 1)

df
```

```python
import numpy as np
```

```python
r = np.arange(1,11)

df = df.set_index(r)

df
```

```python
Top_Ten_ODI_Team = df
Top_Ten_ODI_Team
```
